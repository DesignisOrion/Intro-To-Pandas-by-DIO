## How do I read a tabular data file into pandas?


##### What is Tabular Data?

Tabular data is data that looks like a table, excel spreadsheet. Has rows and columns. Common formats are .csv.


```python
import pandas as pd
```

```python
pd.read_table('http://bit.ly/chiporders') # reading data from a URL
```

```python
# Now let's save it as a dataframe object

orders = pd.read_table('http://bit.ly/chiporders')
```

```python
# call to see the first 5 default
orders.head()
```

```python
pd.read_table('http://bit.ly/movieusers')
```

```python
# Notice That the data has an issue. It placed everything in one column. 
# Do you notice the pipe separator in the data?
# We need to communicate to Pandas of the separator in the data. 
# Read documentation on read_table 'sep'
```

```python
# Result when placing the 'sep' parameter. 
pd.read_table('http://bit.ly/movieusers', sep='|')
```

```python
# Another issue: The header row shows data from the body of the table. 
# In this case, we tell Pandas there isn't a header.
```

```python
# Result when telling Pandas there is no header.
pd.read_table('http://bit.ly/movieusers', sep='|', header=None)
```

```python
# Now look, we need to label our columns. Let's do so.
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code'] # name of columns
pd.read_table('http://bit.ly/movieusers', sep='|', header=None, names=user_cols)
```

```python
# Now lets save it as a DataFrame object: 'users'
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code'] 
users = pd.read_table('http://bit.ly/movieusers', sep='|', header=None, names=user_cols)
```

```python
# then lets call to see the first 5 default.
users.head()
```
